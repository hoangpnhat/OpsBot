```python
sampled_df = df.groupby('problem').apply(lambda x: 
                                         x.sample(n=15, replace=True if len(x) < 10 else False)
                                         ).reset_index(drop=True)
```

```python
# Concatenate the conversations
def concat_conversations(group):
    # Join each list in 'conversation' into a single string and then concatenate all with the specified separator
    return '\n\n\n'.join(group['conversation'].apply(lambda x: '-'*30 + '\n'+ '\n'.join(x)))

# Apply the function to each group
concatenated_conversations = sampled_df.groupby('problem').apply(concat_conversations)
conversation_dict = concatenated_conversations.to_dict()
```

```python
prompt="""Bn l chuyn chuyn vin h tr x l vn  (vit tt l CX) cho cng ty thi trang YODY ti Vit Nam. 
Nhim v ca bn bao gm:
    - Tip nhn vn  t ngi dng, phn loi c vn  trong danh sch cc vn  c lit k sn  di y.
    - Nu khng phn loi c vn , hy yu cu lin h iu phi vin @Coordinators.
    - Nu yu cu ca nhn vin khng r rang, hy yu cu h cung cp thm thng tin.
    -  xut gii php x l nu bn chc chn.

    V qua nhng cuc hi thoi  h tr, bn s a ra 1 quy trnh chun  x l tng vn 
    Quy trnh s c 4 bc:
    - Xc nh cc vn  con (sub-problem) thng gp.
    - Xc nh ngi c th gii quyt c vn , nu r username ca h bng c php @username.
    -  xut gii php x l vn , nu r cc bc cn thc hin.
    - Kim tra mc  hi lng ca ngi dng sau khi gii quyt vn .
    
    #Lu :
    - Chnh sa cc t ng ng chnh t, khng ghi tt.
    - Khng c s dng ting Anh xen k ting Vit\
    - Lu  cc username khng c t vit tt nh Ommi, CNNT, CX, QTRR,... l cc username ca ngi dng trong h thng, khng c kh nng gii quyt vn .


    Di y l cc cuc hi thoi m bn  h tr:
    vn : "{}"
    Cuc hi thoi {}
    
    Quy trnh:
    """
```

```python
client = OpenAI()
```

```python
process_problems = []
for problem,conversation in conversation_dict.items():
    messages = [
        {"role": "system", "content": prompt.format(problem, conversation)},
        {"role": "user", "content": "Hy cho ti quy trnh"},]
    response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=messages
            )
    process_problems.append("problem: " + response.choices[0].message.content)
```

```python
with open('processed_data/process_problems.txt', 'w') as f:
    for item in process_problems:
        f.write("%s\n\n"% item  +"-"*30+'\n\n')
```

```python
with open('processed_data/process_problems.json', 'w') as f:
    json.dump(process_problems, f)
```
