```python
sampled_df = df.groupby('problem').apply(lambda x: 
                                         x.sample(n=15, replace=True if len(x) < 10 else False)
                                         ).reset_index(drop=True)
```

```python
# Concatenate the conversations
def concat_conversations(group):
    # Join each list in 'conversation' into a single string and then concatenate all with the specified separator
    return '\n\n\n'.join(group['conversation'].apply(lambda x: '-'*30 + '\n'+ '\n'.join(x)))

# Apply the function to each group
concatenated_conversations = sampled_df.groupby('problem').apply(concat_conversations)
conversation_dict = concatenated_conversations.to_dict()
```

```python
print(conversation_dict['#van_de: Bảo hành chất liệu'])
```

```python
prompt="""Bạn là chuyên chuyên viên hỗ trợ xử lý vấn đề (viết tắt là CX) cho công ty thời trang YODY tại Việt Nam. 
Nhiệm vụ của bạn bao gồm:
    - Tiếp nhận vấn đề từ người dùng, phân loại được vấn đề trong danh sách các vấn đề được liệt kê sẵn ở dưới đây.
    - Nếu không phân loại được vấn đề, hãy yêu cầu liên hệ điều phối viên @Coordinators.
    - Nếu yêu cầu của nhân viên không rõ rang, hãy yêu cầu họ cung cấp thêm thông tin.
    - Đề xuất giải pháp xử lý nếu bạn chắc chắn.

    Và qua những cuộc hội thoại đã hỗ trợ, bạn sẽ đưa ra 1 quy trình chuẩn để xử lý từng vấn đề
    Quy trình sẽ có 4 bước:
    - Xác định các vấn đề con (sub-problem) thường gặp.
    - Xác định người có thể giải quyết được vấn đề, nêu rõ username của họ bằng cú pháp @username.
    - Đề xuất giải pháp xử lý vấn đề, nêu rõ các bước cần thực hiện.
    - Kiểm tra mức độ hài lòng của người dùng sau khi giải quyết vấn đề.
    
    #Lưu ý:
    - Chỉnh sửa các từ ngữ đúng chính tả, không ghi tắt.
    - Không được sử dụng tiếng Anh xen kẽ tiếng Việt\
    - Lưu ý các username không có từ viết tắt như Ommi, CNNT, CX, QTRR,... là các username của người dùng trong hệ thống, không có khả năng giải quyết vấn đề.


    Dưới đây là các cuộc hội thoại mà bạn đã hỗ trợ:
    vấn đề: "{}"
    Cuộc hội thoại {}
    
    Quy trình:
    """
```

```python
client = OpenAI()
```

```python
process_problems = []
for problem,conversation in conversation_dict.items():
    messages = [
        {"role": "system", "content": prompt.format(problem, conversation)},
        {"role": "user", "content": "Hãy cho tôi quy trình"},]
    response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=messages
            )
    process_problems.append("problem: " + response.choices[0].message.content)
```

```python
with open('processed_data/process_problems.txt', 'w') as f:
    for item in process_problems:
        f.write("%s\n\n"% item  +"-"*30+'\n\n')
```

```python
with open('processed_data/process_problems.json', 'w') as f:
    json.dump(process_problems, f)
```
